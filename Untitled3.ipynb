```python
import pandas as pd

# 列名を手動で指定してデータを読み込む
sakura_df = pd.read_csv('Sakura.csv', header=None, names=['review'])
review_df = pd.read_csv('reviews1.csv', header=None, names=['review'])

# サクラレビューにラベル1、通常レビューにラベル0を付与
sakura_df['label'] = 1
review_df['label'] = 0

# データフレームの結合
df = pd.concat([sakura_df, review_df], ignore_index=True)

# テキストとラベルを抽出
texts = df['review'].tolist()
labels = df['label'].tolist()
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# テキストをトークナイズ
encodings = tokenizer(texts, truncation=True, padding=True, max_length=128, return_tensors='pt')
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# テキストをトークナイズ
encodings = tokenizer(texts, truncation=True, padding=True, max_length=128, return_tensors='pt')
```

```python
import torch
from torch.utils.data import Dataset

class ReviewDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

dataset = ReviewDataset(encodings, labels)
```

```python
from transformers import BertForSequenceClassification

model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir='./results',          # 出力ディレクトリ
    num_train_epochs=3,              # エポック数
    per_device_train_batch_size=8,   # トレーニングバッチサイズ
    per_device_eval_batch_size=8,    # 評価バッチサイズ
    warmup_steps=500,                # ウォームアップステップ
    weight_decay=0.01,               # ウェイト減衰
    logging_dir='./logs',            # ログ出力ディレクトリ
    logging_steps=10,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
)
```

```python
trainer.train()
```

```python
model.save_pretrained("./bert-sakura-model")
tokenizer.save_pretrained("./bert-sakura-model")
```

```python
# テキストの例
text = "ああああああああ"

# トークナイズ
inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)

# 推論
outputs = model(**inputs)
predictions = outputs.logits.argmax(-1)
print(f"Prediction: {'Sakura Review' if predictions.item() == 1 else 'Genuine Review'}")
```
